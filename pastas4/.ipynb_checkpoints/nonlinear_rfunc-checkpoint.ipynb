## Developing a non-linear (time-dependent) response function for Pastas

```python
%matplotlib inline
```

```python
import matplotlib
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 8))

# Make data.
# time
tmax = 100
taumax = 100
t = np.arange(0, tmax, 1)
# lag time at each time step t
tau = np.arange(0, taumax, 1)
t, tau = np.meshgrid(t, tau)
# response
alpha = 0.05
beta = 1
rfunc = np.exp(-alpha * tau) * beta * (np.sin(t/10) + 2)

print(np.shape(rfunc))

# Plot the surface.
surf = ax.plot_surface(t, tau, rfunc, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False, alpha=0.3)

# plot single function evaluations at time t
for time in [i for i in range(0,100,10)]:
    res = (np.exp(-alpha * tau[:, time]) * beta * (np.sin(t[:, time]/10) + 2))
    ax.plot(t[:, time], tau[:, time], res, c="black", lw=3)

ax.set_xlabel("time t")
ax.set_ylabel("lag tau")
ax.set_zlabel("response")
ax.view_init(30, 30)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
